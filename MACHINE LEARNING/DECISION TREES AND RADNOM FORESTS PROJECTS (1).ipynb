# <font color = 'purple'> DEALING WITH CATEGORICAL DATA </font>

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(df,columns = cat_feats,drop_first = True)
```

```python
df.head()
```

# <font color = 'red'> IMPORTING TRAIN TEST SPLIT </font>

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis = 1)

y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

# <font color = 'red'> DECISION TREE </font>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dsc = DecisionTreeClassifier()
```

```python
dsc.fit(X_train,y_train)
```

# <font color ='red'> PREDICTIONS AND EVALUATIONS </font>

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
pred = dsc.predict(X_test)
```

```python
print(classification_report(y_test,pred))
print('\n')
print(confusion_matrix(y_test,pred))
```

# <font color = 'red'> RANDOM FOREST </font>

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 600)
```

```python
rfc.fit(X_train,y_train)
```

```python
preds = rfc.predict(X_test)
```

```python
preds
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

# <font color = 'red'> PREDICTIONS AND EVALUATIONS </font>

```python
print(classification_report(y_test,preds))
print('\n')
print(confusion_matrix(y_test,preds))
```
