```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.pi
```

```python
x = np.arange(0,4 * np.pi,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.title('Sine wave form')
```

```python
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
x
```

```python
y_sin
```

```python
y_cos
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin,c = 'blue',ls = '--',lw = 3)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,c = 'g',ls = '--',lw = 3)
plt.title('Cosine')
```

# <font color = 'rainbow'> PIECHART </font>

```python
labels = 'Prosche Cayenne Turbo S','Range Rover Sport SVR','Toyota Fortuner',' Mercedes GLS 450d'
sizes = [250,350,300,150]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = [0.1,0,0,0]

plt.pie(sizes,labels = labels,colors = colors, explode = explode,
       autopct = '%1.1f%%',shadow = True);
plt.axis('equal');
```
