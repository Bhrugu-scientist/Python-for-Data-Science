```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris);
```

```python
sns.set(style = 'darkgrid')
pg = sns.PairGrid(iris)
pg.map(plt.scatter)
```

```python
pg = sns.PairGrid(iris)

pg.map_diag(sns.distplot)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```
