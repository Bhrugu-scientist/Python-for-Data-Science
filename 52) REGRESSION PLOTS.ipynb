```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set(style = 'darkgrid')
sns.lmplot(y = 'tip',x = 'total_bill',data = tips,hue = 'sex',markers = ['o','v'],scatter_kws = {'s':100})
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips, col = 'sex',row = 'time')
```

```python
sns.lmplot(data = tips, x = 'total_bill',y = 'tip', col = 'day',hue = 'sex', aspect = 0.6, size = 8)
```
