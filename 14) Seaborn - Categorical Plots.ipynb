```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# <font color = 'green'> BARPLOT </font>

```python
sns.barplot(x = 'total_bill',y = 'sex',data = tips)   ## this plot shows the average bill of each sex
```

```python
sns.set(style = 'darkgrid')
sns.barplot(x = 'total_bill',y = 'sex',data = tips, estimator = np.std) ## this plot shows the standard deviation of 
## bill for each sex 
```

```python
sns.countplot(x = 'sex',data = tips)
```

# <font color = 'blue'> BOXPLOT </font>

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = tips, hue = 'smoker')
```

# <font color = 'red'> VIOLIN PLOT </font>

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips,hue = 'sex',split = True)
```

# <font color = 'purple'> STRIP PLOT </font>

```python
sns.stripplot(x = 'day',y = 'total_bill',data = tips,jitter = True,hue = 'sex') 
## jitter will add some random noise to remove the single stack of points.
```

# <font color = 'orange'> SWARM PLOT </font>

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips) 

sns.swarmplot(x = 'day',y = 'total_bill',data = tips,color = 'black')
```

```python
sns.factorplot(x = 'day',y = 'total_bill',data = tips, kind = 'bar')
```
