```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
```

```python
y = x **2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.plot(x,y, 'r-')
plt.show()
```

```python
plt.plot(x,y,'g-')
```

```python
plt.plot(x,y,'-y')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('HAHAHAHAHAH')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('y label')
axes.set_title('T I T L E')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.set_title('S M A L L    I N C E P T I O N')
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('L A R G E    I N C E P T I O N')
```
